## Exercise 2
The amount of uncertainty will not reduce up to a point when the data we observe is biased (specifically selection bias).

For instance, let’s say you want to assess a program for improving the eating habits of shift workers. You put up flyers where many work night shifts and invite them to participate. However, those who sign up may be very different from those who don’t. They may be more health conscious to begin with, which is why they are interested in a program to improve eating habits.

This leads to the fact that the program would be obviously effective as the health of those who took part in was already better than the health of those who didn't.

The program may be not that much effective, but the more and more volunteers you invite would just keep showing that the program was a success.

There are many types of selection biases that would falsify the law of large number. Try to learn about thoses and be careful on your studies.

## Exercise 3

```python
from d2l import mxnet as d2l
from mxnet import np, npx
from mxnet.numpy.random import multinomial

npx.set_np()

SIZE = 10000
npx.random.seed(42)
counts = multinomial(1, [0.5, 0.5], size=SIZE)
cum_counts = counts.astype(np.float32).cumsum(axis=0)
estimates = cum_counts / cum_counts.sum(axis=1, keepdims=True)
```

```python
variance = [0.0] * SIZE
for i in range(1, SIZE):
    variance[i] = float(
        (cum_counts[i, 0] * (0 - estimates[i, 0])**2 + cum_counts[i, 1] *
         (1 - estimates[i, 1])**2) / i)
```

```python
d2l.set_figsize((4.5, 3.5))
d2l.plt.plot(variance, label=("Var(coin)"))
d2l.plt.axhline(y=0.25, color='red', linestyle='dashed')
```

The variance converge to $0.25$ over time.

With the dependence of variance onto mean, it can be implied that the central limit theorem also works with variances (not just means).
